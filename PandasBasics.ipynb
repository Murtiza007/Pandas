```python
#concatination vs merging

section_APrimary={
    'name':['raj','simran','rahul'],
    'science':[90,80,70],
    'maths':[80,90,60],
    'urdu':[70,60,80]
}
section_ASecondary={
    'name':['raj','simran','rahul'],
    'history':[90,80,70],
    'geo':[80,90,60],
    'civics':[70,60,80]
}


section_BPrimary={
    'name':['moin','ahmad','basit'],
    'science':[90,80,70],
    'maths':[80,90,60],
    'urdu':[70,60,80]
}
section_BSecondary={
    'name':['moin','ahmad','basit'],
    'history':[90,80,70],
    'geo':[80,90,60],
    'civics':[70,60,80]
}
 
 # a overall
 # a & b combined
 # b overall
```

```python
A_pry=pd.DataFrame(section_APrimary)
A_sec=pd.DataFrame(section_ASecondary)
B_sec=pd.DataFrame(section_BSecondary)
B_pry=pd.DataFrame(section_BPrimary)

A_pry
A_sec


```

```python
A_overall=pd.merge(A_pry,A_sec,on="name")
A_overall
```

```python
B_overall=pd.merge(B_pry,B_sec,on="name")
B_overall
```

```python
sec_overall=pd.concat([A_overall,B_overall])
sec_overall
sec_overall.reset_index(inplace=True)
sec_overall
sec_overall.pop("index")
sec_overall
```
